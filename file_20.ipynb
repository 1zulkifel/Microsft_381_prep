```python
def exp(base,power):#method for exponentials
    return base ** power
s=exp(2,3)
print(s)
#or 
def two_to(power):# method better than above
    return exp(2,power)
rslt=two_to(2)
print(rslt)
#or
from functools import partial#better than above all
two_to_the=partial(exp,3)
print(two_to_the (2))
```

```python
import numpy as np
import cv2

cap = cv2.VideoCapture(0)
 
while(True):
    ret, frame = cap.read()
    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
 
    cv2.imshow('frame',gray)
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```

```python
#creating a gui
import tkinter as tk

window=tk.Tk()
txt_box=tk.Entry(window)
txt_box2=tk.Entry(window)

def SaveText():
    str_1=txt_box.get()
    fx=open('file_1','w+')
    fx.write(str_1)
    fx.close()
    str_2=txt_box2.get()
    ps=open('fiel_2','W+')
    ps.write(str_2)
    ps.close()
btn_1=tk.Button(window,text=('save'),command=SaveText)
txt_box.pack()
btn_1.pack()
btn_2=tk.Button(window,text=('cancel'),command=SaveText)
txt_box2.pack()
btn_2.pack()
window.mainloop()
```
